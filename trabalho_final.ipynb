```python
###Resultados:
##Por meio do gráfico de investimentos do BNDES por setor conseguimos perceber que há uma escalada de crescimento nos investimentos totais do BNDES à partir do ano de 2008 mantendo o crescimento até metade da segunda década dos anos 2000. Esta escalada de crescimento dar-se principalmente por meio dos dispêndios com os setores de comercio e serviços seguido da indústria de transformação . A agropecuária teve um crescimento discreto com uma pequena elevação no de 2014 , mantendo praticamente constante  para os anos seguintes. Dentre os quatro setores analisados, Indústria extrativa , Comercio e Serviços e Agropecuária dos e Industria de transformação, a indústria extrativa foi a que representou o menor crescimento ,os dispêndios para esse setor manteve-se praticamente constante ao longo dos 27 anos. Do ano de 2015 até 2021 houve uma queda dos investimentos retornando o crescimento no ano de 2022.
##Os dispêndios, na média, mantem-se sem grandes variações do mês de  fevereiro até o mês de novembro com crescimento no mês de dezembro  e queda significativa no mês de janeiro. Essa tendencia é perceptível por meio da análise do gráfico de Sazonalidade. O comportamento é valido para todos os setores analisados respeitando os níveis de investimentos por setor analisados no gráfico anterior.
##No período da pandemia do Covid-19 os dispêndios sofreram duas quedas significativas, a primeira em janeiro de 2021 e a segunda em janeiro de 2022.Lembro que as restrições no período da pandemia iniciaram em março de 2020 e que o mês de janeiro , como foi observado, é um mês de redução dos investimentos do BNDES.

```
